```python
from google_images_download import google_images_download

politicians = ["Donald Trump", "Barack Obama", "Michelle Obama", "Joe Biden", "Ivanka Trump", "Melania Trump"]
output_directory = "/home/henning/git_repos/htwg_mldl/Versuche/data/PoliticiansFace/downloads"

for politician in politicians:
    response = google_images_download.googleimagesdownload()
    arguments = {"keywords":politician, "limit":5000, "print_urls":False, "chromedriver":"/usr/bin/chromedriver", "format": "jpg", "type": "face", "output_directory": output_directory}
    paths = response.download(arguments)
```

```python
import cv2
import cvlib as cv
import glob
import os
import numpy as np

X = []
Y = []
root_path = "/home/henning/git_repos/htwg_mldl/Versuche/data/PoliticiansFace"
for politician in politicians:
    path = os.path.join(root_path, "downloads", politician, "*")
    files = glob.glob(path)
    for file in files:
        img = cv2.imread(file)
        
        if img is None: continue
        
        faces, confidences = cv.detect_face(img)
        for face in faces:
            (startX,startY) = face[0],face[1]
            (endX,endY) = face[2],face[3]
            face_img = img[startY:endY, startX:endX]  
            
            if face_img.shape[0] == 0 or face_img.shape[1] == 0: continue
            
            face_img = cv2.resize(face_img, (256, 256))
            X.append(face_img)
            Y.append(politician)
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
print(X.shape, Y.shape)
```

```python
for politician in politicians:
    print("{}: {}".format(politician, np.sum(Y == politician)))
```

```python
np.save(os.path.join(root_path, "just_faces", "X.npy"), X)
np.save(os.path.join(root_path, "just_faces", "Y.npy"), Y)
```
