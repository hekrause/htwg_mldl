```python
import cv2
import numpy as np
import glob
```

```python
path = '/home/henning/tmp/donald_trump_faces/*.jpg'
files = glob.glob(path)   

cv2.destroyAllWindows()

n = len(files)
good_files = []
for i in range(n):
    image_path = files[i]
    img = cv2.imread(image_path)

    if img is not None:
        img = cv2.resize(img, (256, 256))
        good_files.append(img)
```

```python
path = '/home/henning/tmp/donald_trump_faces/*.jpg'
files = glob.glob(path)   

cv2.destroyAllWindows()

# X = np.zeros(shape=(len(good_files), 256, 256, 3), dtype=np.uint8)
# Y = np.zeros(shape=(len(good_files)), dtype=np.uint8)
X = []
Y = []
trump = 0
other = 0
n = 3000
for i in range(n):
    image_path = files[i]
    img = cv2.imread(image_path)
    
    if img is not None:
        img = cv2.resize(img, (256, 256))
        cv2.imshow("img", img)
        key = cv2.waitKey(0)

        if key == ord('t'): # trump
            X.append(img)
            Y.append(1)
            trump += 1
            cv2.destroyAllWindows()
        else: # other
            X.append(img)
            Y.append(0)
            other += 1
            cv2.destroyAllWindows()

        if (i % 10) == 0:
            print("imageidx: {}, trump: {}, other: {}".format(i, trump, other))
            X_tmp = np.asarray(X)
            Y_tmp = np.asarray(Y)
            print(X_tmp.shape, Y_tmp.shape)
            np.save("/home/henning/git_repos/htwg_mldl/Versuche/data/TrumpFace/X.npy", X_tmp)
            np.save("/home/henning/git_repos/htwg_mldl/Versuche/data/TrumpFace/Y.npy", Y_tmp)
            cv2.destroyAllWindows()
cv2.destroyAllWindows()
```
